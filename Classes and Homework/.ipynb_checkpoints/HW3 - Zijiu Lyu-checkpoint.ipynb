```python
import numpy as np
import pandas as pd
import pandas_datareader as pd_data
import statsmodels.api as sm
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error as MSE
from sklearn.linear_model import Lasso, Ridge
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import arma_order_select_ic
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.gofplots import qqplot
from statsmodels.nonparametric.smoothers_lowess import lowess
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sn
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = [8, 6]
```

# Problem 1 - GS Daily Log Returens

```python
start_date = '2018-01-01'
end_date = '2018-12-31'
idx = ['GS', 'SPY', 'DIA', 'WMT', 'EURUSD=X', 'GBPUSD=X']
channel = 'yahoo'

data = pd.DataFrame(index=pd.date_range(start_date, end_date))
data.index.name = 'Date'
for stock in idx:
    stock_data = pd_data.DataReader(stock, channel, start_date,
                                    end_date)['Adj Close']
    stock_data.name = stock
    data = pd.merge(left=data,
                    right=stock_data,
                    how='left',
                    left_index=True,
                    right_index=True)
    
data.dropna(inplace=True)
data = np.log(data).diff().dropna()
data.head()
```

## (a) OLS

```python
y = data['GS']
X = data.drop('GS', axis=1)
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

## (b) OLS CV MSE

```python
y = data['GS']
X = data.drop('GS', axis=1)
X = sm.add_constant(X)
y_pred = pd.Series(index=data.index, name='GS pred')

k_fold = KFold(n_splits=5, shuffle=True, random_state=2022)
for train_idx, test_idx in k_fold.split(data):
    y_fold = y.iloc[train_idx]
    X_fold = X.iloc[train_idx]
    model_fold = sm.OLS(y_fold, X_fold).fit()
    y_pred.iloc[test_idx] = model_fold.predict(X.iloc[test_idx])

mse_ols = MSE(y, y_pred)
print(f'5 Fold MSE: {mse_ols}')
```

## (c) LASSO

```python
y = data['GS']
X = data.drop('GS', axis=1)
k_fold = KFold(n_splits=5, shuffle=True, random_state=2022)

mse_lasso = []
range_lasso = np.logspace(-8, -4, 20)
for lmd in range_lasso:
    y_pred = pd.Series(index=data.index, name='GS pred')
    for train_idx, test_idx in k_fold.split(data):
        y_fold = y.iloc[train_idx]
        X_fold = X.iloc[train_idx]
        model_fold = Lasso(alpha=lmd,
                           fit_intercept=True,
                           normalize=True,
                           random_state=2022).fit(X_fold, y_fold)
        y_pred.iloc[test_idx] = model_fold.predict(X.iloc[test_idx])
    mse_lasso.append(MSE(y, y_pred))
argmin_lasso = np.argmin(mse_lasso)
print(f'Best Lambda: {range_lasso[argmin_lasso]}')
print(f'Min MSE: {mse_lasso[argmin_lasso]}')
```

```python
plt.plot(range_lasso, mse_lasso)
plt.title('Lasso Regression', fontsize=15)
plt.xlabel('Lambda', fontsize=15)
plt.ylabel('MSE', fontsize=15)
plt.scatter(range_lasso[argmin_lasso],
            mse_lasso[argmin_lasso],
            label='Optimal Point',
            color='g')
plt.hlines(mse_ols,
           range_lasso[0],
           range_lasso[-1],
           linestyles='--',
           color='r',
           label='OLS')
plt.legend(fontsize=15)
plt.savefig('figs/lasso.png')
plt.show()
```

## (d) Ridge

```python
y = data['GS']
X = data.drop('GS', axis=1)
k_fold = KFold(n_splits=5, shuffle=True, random_state=2022)

mse_ridge = []
range_ridge = np.linspace(0.0005, 0.05, 20)
for lmd in range_ridge:
    y_pred = pd.Series(index=data.index, name='GS pred')
    for train_idx, test_idx in k_fold.split(data):
        y_fold = y.iloc[train_idx]
        X_fold = X.iloc[train_idx]
        model_fold = Ridge(alpha=lmd,
                           fit_intercept=True,
                           normalize=True,
                           random_state=2022).fit(X_fold, y_fold)
        y_pred.iloc[test_idx] = model_fold.predict(X.iloc[test_idx])
    mse_ridge.append(MSE(y, y_pred))
argmin_ridge = np.argmin(mse_ridge)
print(f'Best Lambda: {range_ridge[argmin_ridge]}')
print(f'Min MSE: {mse_ridge[argmin_ridge]}')
```

```python
plt.plot(range_ridge, mse_ridge)
plt.title('Ridge Regression', fontsize=15)
plt.xlabel('Lambda', fontsize=15)
plt.ylabel('MSE', fontsize=15)
plt.scatter(range_ridge[argmin_ridge],
            mse_ridge[argmin_ridge],
            label='Optimal Point',
            color='g')
plt.hlines(mse_ols,
           range_ridge[0],
           range_ridge[-1],
           linestyles='--',
           color='r',
           label='OLS')
plt.legend(fontsize=15)
plt.show()
```

## (f) Optimal LASSO

```python
y = data['GS']
X = data.drop('GS', axis=1)

model = Lasso(alpha=range_lasso[argmin_lasso],
              fit_intercept=True,
              normalize=True,
              random_state=2022).fit(X, y)
print(model.coef_)
print([model.intercept_])
```

# Problem 2 - Manufacturing Productivity in the UK

## 数据预处理和可视化

```python
data = pd.read_csv('mfp.csv', index_col=0)

index_new = []
for date in data.index:
    tmp = date[:4]
    if date[-1] == '1':
        tmp += '-03'
    elif date[-1] == '2':
        tmp += '-06'
    elif date[-1] == '3':
        tmp += '-09'
    else:
        tmp += '-12'
    index_new.append(tmp)
data.index = pd.to_datetime(index_new)
data.head()
```

```python
plt.plot(data.index, data)
plt.title('Productivity Index', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.savefig('figs/prod.png')
plt.show()
```

```python
data = data.diff().dropna()

plt.plot(data.index, data)
plt.title('Productivity Index Diff', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.savefig('figs/prod diff.png')
plt.show()
```

## 平稳性检验

```python
adfuller(data.values)
```

```python
kpss(data.values)
```

所以growth本身可以视为平稳

## RA(2)

```python
best_model = arma_order_select_ic(data.values,
                                  max_ar=10,
                                  max_ma=0,
                                  ic=['aic', 'bic'])
best_model
```

```python
model = ARIMA(data.values, order=(2, 0, 0), trend='c').fit()
print(model.summary())
```

```python
qqplot(model.resid, fit=True, line='45')
plt.savefig('figs/qqplot.png')
plt.show()
```

```python
plot_acf(model.resid, zero=False)
plt.savefig('figs/resid acf.png')
plt.show()
```

```python
plot_pacf(model.resid, zero=False)
plt.savefig('figs/resid pacf.png')
plt.show()
```

```python
lags = np.floor(np.log(len(data)))
ljbox_test_res_ciscoAR1 = sm.stats.acorr_ljungbox(model.resid, lags)
plt.plot(ljbox_test_res_ciscoAR1['lb_pvalue'])
plt.axhline(y=0.05, color='r', ls='--')
plt.title('LB Test', fontsize=15)
plt.xlabel('Lag', fontsize=15)
plt.ylabel('P-value', fontsize=15, rotation=0)
plt.savefig('figs/lbtest.png')
plt.show()
```

```python
pred = []
for i in range(len(data), len(data) + 10):
    pred.append(model.predict(i)[0])

pred_index = [
    '2018-09-01', '2018-12-01', '2019-03-01', '2019-06-01', '2019-09-01',
    '2019-12-01', '2020-03-01', '2020-06-01', '2020-09-01', '2020-12-01'
]
pred = pd.DataFrame(index=pd.to_datetime(pred_index), data=pred)
pred.columns = ['MFP']

tmp = pd.concat([data, pred], axis=0)
plt.plot(tmp.index, tmp.values, label='Exact')
plt.plot(pred.index, pred.values, label='Pred')
avg = lowess(tmp.values.reshape(-1),
             range(len(tmp)),
             return_sorted=False,
             frac=0.4)
plt.plot(tmp.index, avg, label='Non-param Fit')
plt.title('Productivity Growth Pred', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.hlines(0, tmp.index[0], tmp.index[-1], ls='--', color='r')
plt.legend(fontsize=15)
plt.savefig('figs/prod pred.png')
plt.show()
```

## 两个set上re-fit

```python
data_1 = data[data.index >= '1997']
data_1 = data_1[data_1.index < '2008']
model_1 = ARIMA(data_1.values, order=(2, 0, 0)).fit()
print(model_1.summary())
```

```python
data_2 = data[data.index >= '2008']
model_2 = ARIMA(data_2.values, order=(2, 0, 0)).fit()
print(model_2.summary())
```

```python
for i in range(4):
    x_1 = model_1.params[i]
    x_2 = model_2.params[i]
    s_1 = model_1.cov_params()[i, i]
    s_2 = model_2.cov_params()[i, i]
    d_1 = len(data_1) - 4
    d_2 = len(data_2) - 4

    t = (x_1 - x_2) / np.sqrt(s_1 + s_2)
    df = (s_1 + s_2)**2 / (s_1**2 / d_1 + s_2**2 / d_2)
    print(model_1.param_names[i], ':', 2 * scipy.stats.t.sf(abs(t), df=df))
```

```python
pred = []
for i in range(len(data_2), len(data_2) + 10):
    pred.append(model_2.predict(i)[0])

pred_index = [
    '2018-09-01', '2018-12-01', '2019-03-01', '2019-06-01', '2019-09-01',
    '2019-12-01', '2020-03-01', '2020-06-01', '2020-09-01', '2020-12-01'
]
pred = pd.DataFrame(index=pd.to_datetime(pred_index), data=pred)
pred.columns = ['MFP']

tmp = pd.concat([data_2, pred], axis=0)
plt.plot(tmp.index, tmp.values, label='Exact')
plt.plot(pred.index, pred.values, label='Pred')
plt.title('Productivity Growth Pred', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.hlines(0, tmp.index[0], tmp.index[-1], ls='--', color='r')
plt.legend(fontsize=15)
plt.savefig('figs/prod pred new.png')
plt.show()
```

# Problem 3 - GOOG

## 数据预处理

```python
start_date = '2012-01-04'
end_date = '2019-11-04'
stock = 'GOOG'
channel = 'yahoo'

data = pd_data.DataReader(stock, channel, start_date, end_date)['Volume']
data.head(5)
```

```python
mean = data.mean()
std = data.std()
for i in data.index:
    if data.loc[i] > mean + 3 * std or data.loc[i] < mean - 3 * std:
        data.drop(i, inplace=True)

plt.plot(data.index, data)
plt.title('Google', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Vol', fontsize=15)
plt.savefig('figs/goog.png')
plt.show()
```

```python
adfuller(data.values)
```

```python
data = data.diff().dropna()

plt.plot(data.index, data)
plt.title('Google', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Vol Diff', fontsize=15)
plt.savefig('figs/goog diff.png')
plt.show()

data.head()
```

```python
adfuller(data.values)
```

```python
kpss(data.values)
```

## ARIMA(3, 1, 1)

注意，此部分与3.2为并列关系，在进行此部分之前不要对data进行diff操作，直接使用下载下来的原始数据即可。

```python
data /= 1e8
```

```python
model = ARIMA(data.values, order=(3, 1, 1)).fit()
print(model.summary())
```

```python
pred_index = pd.to_datetime(['2019-11-05', '2019-11-06', '2019-11-07', '2019-11-08'])
forecast = model.get_forecast(4)
pred_mean = forecast.predicted_mean
pred_lb = forecast.conf_int(alpha=0.05)[:, 0]
pred_ub = forecast.conf_int(alpha=0.05)[:, 1]

pred = pd.Series(index=pred_index, data=pred_mean)
tmp = pd.concat([data[data.index >= '2019-10-01'], pred], axis=0)
plt.plot(tmp.index, tmp.values, label='Exact')
plt.plot(pred.index, pred.values, label='Pred')
plt.plot(pred.index, pred_lb, label='95% lb')
plt.plot(pred.index, pred_ub, label='95% ub')
plt.title('Google Vol Pred (Unit: 1e8)', fontsize=15)
plt.xlabel('Time', fontsize=15)
plt.legend(fontsize=15)
plt.savefig('figs/goog pred.png')
plt.show()
```
