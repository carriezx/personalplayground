```python
#daily log return
allData_log_return = np.log(allData).diff().dropna()
allData_log_return.head()
```

```python
allData_log_return.shape
```

```python
allData_log_return.hist(bins=100,figsize=(12,4),color='b', density = True)
```

```python
plt.hist(allData_log_return['TSLA'],bins=100, density = True)
mu1, std1 = stats.norm.fit(allData_log_return['TSLA']) 
x = np.linspace(-0.25, 0.25, 100)
best_fit_line = stats.norm.pdf(x, mu1, std1)
plt.title('TSLA compare with the normal probability densities')
plt.plot(x, best_fit_line)
```

```python
plt.hist(allData_log_return['GS'],bins=100, density = True)
mu2, std2 = stats.norm.fit(allData_log_return['GS']) 
x = np.linspace(-0.2, 0.2, 100)
best_fit_line = stats.norm.pdf(x, mu2, std2)
plt.title('GS compare with the normal probability densities')
plt.plot(x, best_fit_line)
```

```python
import pandas_datareader as pd_data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
import math
import scipy as sp
```

```python
FuelData = pd.read_csv('FuelData.csv')
FuelData.head()
```

```python
ConvFuel = pd.DataFrame(FuelData['City FE (Guide) - Conventional Fuel'])
ConvFuel.hist(bins = 50, figsize = (15,10), color = 'm', density = True)
```

```python
ConvFuel['Mfr Name'] = FuelData['Mfr Name']
ConvFuel['Mfr Name'].value_counts()
```

```python
ConvFuel['Mfr Name'] = ConvFuel['Mfr Name'].astype('category')
```

```python
ConvFuel.boxplot(by = 'Mfr Name', rot=45, fontsize = 10, figsize = (12,8))
```

```python
ConvFuel.head(10)
```

```python
#Least fuel efficient 
idx = ConvFuel['City FE (Guide) - Conventional Fuel'].idxmin()
ConvFuel.loc[idx]['Mfr Name']
```

```python
#Most fuel efficient 
idx = ConvFuel['City FE (Guide) - Conventional Fuel'].idxmax()
ConvFuel.loc[idx]['Mfr Name']
```

```python
plt.title('Fuel efficiency against the engine displacement')
plt.scatter(FuelData['Eng Displ'], ConvFuel['City FE (Guide) - Conventional Fuel'])
plt.xlabel('Eng Displ')
plt.ylabel('Fuel Efficiency')
```

```python
plt.title('Fuel efficiency against the number of cylinders')
plt.scatter(FuelData['# Cyl'], ConvFuel['City FE (Guide) - Conventional Fuel'])
plt.xlabel('#Cyl')
plt.ylabel('Fuel Efficiency')
```

```python
#Regress FE on Engine Displacement

import statsmodels.api as sm
x = FuelData['Eng Displ']
y = FuelData['City FE (Guide) - Conventional Fuel']

Xmod = sm.add_constant(x)

model = sm.OLS(y, Xmod).fit()
y_fitted = model.predict(Xmod)

model.summary()
```

```python
#plot of FE vs Eng Disp Linear Regression; Linearity = Failed

fig = plt.figure(figsize = (6,5))
plt.plot(x, y_fitted, color = 'r', linewidth = 3)
plt.scatter(FuelData['Eng Displ'], ConvFuel['City FE (Guide) - Conventional Fuel'])
plt.title('FE vs Eng Disp with Linear Regression', size = 15)
plt.xlabel('Engine Displacement')
plt.ylabel('Fuel Efficiency')
plt.show()
```

```python
#Residuals do not have saame variance; Homoskedasticity - Failed

import seaborn as sns
plt.plot(figsize = (6,5))
res = model.resid
sns.regplot(y_fitted, res, lowess = False, scatter_kws = {'alpha': 0.5}, line_kws = {'color': 'r'})
plt.xlabel('y fitted values')
plt.ylabel('residuals')
```

```python
#Normality Test -- QQ plot

sm.qqplot(res, fit = True, line = '45')
plt.xlabel('residual quantiles')
plt.ylabel('normal quantiles')
plt.show()
```

```python
#Regress log(FE) on Engine Displacement

x = FuelData['Eng Displ']
logy = np.log(FuelData['City FE (Guide) - Conventional Fuel'])

Xmod = sm.add_constant(x)

model_log = sm.OLS(logy, Xmod).fit()
ylog_fitted = model_log.predict(Xmod)

model_log.summary()
```

```python
#plot of log(FE) vs Eng Disp Linear Regression

plt.figure(figsize = (6,5))
plt.plot(x, ylog_fitted, color = 'r', linewidth = 3)
plt.scatter(x, logy)
plt.title('log(FE) vs Eng Disp with Linear Regression', size = 15)
plt.xlabel('Engine Displacement')
plt.ylabel('log Fuel Efficiency')
plt.show()
```

```python
res_log = model_log.resid
sns.regplot(ylog_fitted, res_log, lowess = False, scatter_kws = {'alpha': 0.5}, line_kws = {'color': 'r'})
plt.xlabel('y fitted values')
plt.ylabel('residuals')
```

```python
#Normality test QQplot

sm.qqplot(res_log, fit = True, line = '45')
plt.xlabel('residual quantiles')
plt.ylabel('normal quantiles')
plt.show()
```

Does adding z as a parameter improve the model?

```python
Autolist = []
N = len(y)
for i in range(N):
    if FuelData['Transmission'][i].startswith('A'):
        Autolist.append(1)
    else:
        Autolist.append(0)
FuelData['Autolist'] = Autolist
FuelData['Autolist'].value_counts()
```

```python
#Linear Regression with z

x = FuelData['Eng Displ']
logy = np.log(FuelData['City FE (Guide) - Conventional Fuel'])

Xnew = np.column_stack((x.values, FuelData['Autolist'].values))
Xmod = sm.add_constant(Xnew)

model2 = sm.OLS(logy,Xmod).fit()
predictions = model2.predict(Xmod)
res = model2.resid

model2.summary()
```

```python
AutoX = Xmod[FuelData['Autolist'] == 1]
AutoY = model2.predict(AutoX)

ManX = Xmod[FuelData['Autolist'] == 0]
ManY = model2.predict(ManX)
```

```python
plt.scatter(x,logy)
plt.plot(AutoX[:,1], AutoY, color = 'r', linewidth = 3)
plt.plot(ManX[:,1], ManY, color = 'g', linewidth = 3)
plt.title('FE vs Eng Disp regarding Auto type')
plt.xlabel('Engine Displacement')
plt.ylabel('log Fuel Efficiency')
```

```python
#Add z variable using another method 'hue' argument

sns.lmplot(x = 'Eng Displ', y = 'City FE (Guide) - Conventional Fuel', data = FuelData, fit_reg = True, hue = 'Autolist', scatter_kws = {'color':'r'})
```

```python
#Largest residual fuel efficient 
idx = res.idxmax()
FuelData.loc[idx]['Mfr Name']
```

Adding quadratic x as a predictor for linear regression

```python
x = FuelData['Eng Displ']
z = x ** 2
logy = np.log(FuelData['City FE (Guide) - Conventional Fuel'])

Xsqr = np.column_stack((x.values, z.values))
Xmodsqr = sm.add_constant(Xsqr)


model_square = sm.OLS(logy, Xmodsqr).fit()
prediction_sqr = model_square.predict(Xmodsqr)
resid_sqr = model_square.resid

model_square.summary()
```

```python
dic = {'X': x, 'Y': prediction_sqr}
df_sqr = pd.DataFrame(dic, columns = ['X', 'Y'])
df_sqr.sort_values(by = ['X'], inplace = True)
```

```python
plt.figure(figsize = (6,5))

plt.plot(df_sqr['X'], df_sqr['Y'], color = 'r', linewidth = 3)
plt.scatter(x, logy)
plt.title('log(FE) vs Eng Disp with quadratic Linear Regression', size = 15)
plt.xlabel('Engine Displacement')
plt.ylabel('log Fuel Efficiency')
```

```python
#Normality test for x square via QQ plot

sm.qqplot(resid_sqr, fit = True, line = '45')
plt.title('qq plot of linear regression with x^2')
plt.xlabel('residual quantiles')
plt.ylabel('normal quantiles')
plt.show()
```
